```python
#Importing the page into a beautifulsoup object called 'soup'
#url_julio_nfl = 'https://www.pro-football-reference.com/players/J/JoneJu02.htm'
url_julio_nfl = 'https://www.sports-reference.com/cfb/players/julio-jones-1.html'
html =urlopen(url_julio_nfl)
soup = BeautifulSoup(html,'html.parser')
```

```python
#If you open the page on your web browser, there are multiple tables. 
#Only a few are of interest, but findAll seems to return only 2.
table=soup.findAll('table')
len(table)
```

```python
new_table = pd.DataFrame(columns=['year','school','conference','class','pos','games',
                                  'receptions','yards', 'average','td',
                                 'attemps rushing','yards rushing','avg rushing','td rushing',
                                 'scrimmages','yards total','avg total','td total'],
                         index=range(0,10)) # I know the size 
    
row_marker = 0
for row in table[0].find_all('tr')[2:]:
    column_marker = 0
    columns = row.find_all('td')
    for column in columns:
        new_table.iat[row_marker,column_marker] = column.get_text()
        column_marker += 1
    row_marker += 1
    
new_table
```
