```python
%matplotlib inline
```

```python
import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf

# fix error in some versions of statsmodels
import scipy
if not hasattr(scipy.stats,'chisqprob'):
    scipy.stats.chisqprob = lambda chisq, df: scipy.stats.chi2.sf(chisq, df)
```

```python
smoking = pd.read_csv("whickham.csv")
```

```python
smoking.info()
```

```python
smoking = smoking[smoking.age < 65]
```

```python
smoking.outcome = 1*(smoking.outcome == 'Alive')
```

```python
smoking.outcome.head()
```

```python
pp.scatter(smoking.age,smoking.outcome)
```

```python
color = np.where(smoking.smoker == 'No','skyblue','coral')
jitter = 0.05*(np.random.rand(len(smoking)) - 0.5)
```

```python
pp.scatter(smoking.age,smoking.outcome+jitter,c=color,linewidths=0.5,s=50,edgecolor='k',alpha=0.5)
```

```python
ols_group = smf.ols('outcome ~ smoker',data=smoking).fit()
```

```python
def plotfit(fit):
    pp.scatter(smoking.age,smoking.outcome+jitter,c=color,linewidths=0.5,s=50,edgecolor='k',alpha=0.5)
    pp.scatter(smoking.age,fit.predict(smoking),c=color,linewidths=0.5,s=50,edgecolor='gray',marker='d')
```

```python
plotfit(ols_group)
```

```python
ols_group.mse_resid
```

```python
ols_group.params
```

```python
sm.stats.anova_lm(ols_group)
```

```python
ols_group.conf_int()
```

```python
ols_group_age = smf.ols('outcome ~ smoker + age',data=smoking).fit()
```

```python
plotfit(ols_group_age)
```

```python
ols_group_age.mse_resid
```

```python
ols_group_age.params
```

```python
logit_smoker_age = smf.logit('outcome ~ smoker + age',data=smoking).fit()
```

```python
plotfit(logit_smoker_age)
```

```python
logit_smoker_age.llf
```

```python
logit_smoker_age.params
```

```python
logit_smoker_age.conf_int()
```

```python
logit_smoker_age.summary()
```
